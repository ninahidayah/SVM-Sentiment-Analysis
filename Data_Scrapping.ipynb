```python
app_bca = 'com.bca'
app_info = app(app_bca)
```

```python
# country indonesia
reviews = reviews_all(
    app_bca,
    lang='id',
    country='id'
)
```

```python
# Create a DataFrame from the review results
dataframe = pd.DataFrame(reviews)
```
